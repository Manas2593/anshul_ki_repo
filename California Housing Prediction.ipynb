# Get the Data

```python
# Downloading the data
from sklearn.datasets import fetch_california_housing
def fetch_housing_data():
    data=fetch_california_housing(as_frame=True)
    return data
```

```python
# Loading the dataset
data=fetch_housing_data()
```

```python
# Description of the dataset
print(data.DESCR)
```

```python
X = data.data
X
```

```python
import pandas as pd
Y = pd.Series(data=data.target, name='MedHouseVal')
Y
```

```python
df=pd.concat([X,Y],axis=1)
df
```

```python
# Checking for non-null values
df.info()
```

```python
# Describing the dataset
df.describe()
```

```python
# PLotting histogram of params
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
(df.hist(bins=50,figsize=(20,15)))
plt.show()
```

```python
# Splitting Dataframe into test and train case
import numpy as np
def split_test_data(df,test_split_ratio):
    random_permutation_indices = np.random.permutation(len(df))
    test_df_size = int(len(df)*test_split_ratio)
    test_split_indices = random_permutation_indices[:test_df_size]
    train_split_indices = random_permutation_indices[test_df_size:]
    return df.iloc[train_split_indices], df.iloc[test_split_indices]
```

```python
# Calling the test_train custom function
train_df,test_df=split_test_data(df,0.2)
```

```python
train_df.shape, test_df.shape
```

```python
#  We want equal representation of all income categories in our test data. Let;'s divide data into income categories
df["income_cat"]=pd.cut(df["MedInc"],bins=[0,1.5,3,4.5,6.,np.inf],labels=[1,2,3,4,5])
df["income_cat"].hist()
plt.show()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
def strat_split(df):
    stratified_split=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
#     stratified_split
    for train_index,test_index in stratified_split.split(df,df["income_cat"]):
        strat_train_set=df.loc[train_index]
        strat_test_set=df.loc[test_index]
    
    return strat_train_set,strat_test_set
```

```python
strat_train, strat_test = strat_split(df)
```

```python
strat_train.shape, strat_test.shape
```

```python
# Let's see if our output matches our expectations
strat_train.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
# Removing income_cat from both sets
strat_train.drop("income_cat",axis=1,inplace=True)
strat_test.drop("income_cat",axis=1,inplace=True)
```
